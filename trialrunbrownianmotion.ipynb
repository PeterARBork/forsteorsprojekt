```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownmotion(mu,sigma,N):
    return np.cumsum(np.random.normal(mu,sigma,N))
```

```python
mu=0
sigma=1
N=1000
plt.plot(brownmotion(mu,sigma,N),brownmotion(mu,sigma,N))
```
